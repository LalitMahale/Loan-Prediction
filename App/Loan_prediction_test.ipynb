```python
col = [ 'Dependents', 'Education',
       'Self_Employed', 
       'Loan_Amount_Term', 'Credit_History', 'Property_Area']

for i in col:
    print(test[i].unique())
```

```python
test.columns
```

```python

test["Dependents"] = test["Dependents"].map({"0":0,"1":1,"2":2,"3+":3})

test["Education"] = test["Education"].map({"Graduate":1,"Not Graduate":0})

test["Self_Employed"] = test["Self_Employed"].map({"Yes":1,"No":0})

test["Property_Area"] = test["Property_Area"].map({"Urban":1,"Rural":0, "Semiurban":2})
```

```python
test.head()
```

```python
test.isnull().sum()
```

```python
"LoanAmount"
```

```python
lst = ["Dependents","Self_Employed","Credit_History","Loan_Amount_Term","Property_Area"]

for i in lst:
    test[i] = test[i].fillna(test[i].mode()[0])
```

```python
test["LoanAmount"] = test["LoanAmount"].fillna(round(test["LoanAmount"].mean()))
```
